```python
s=input("Enter main string")
substr=input("Enter sub string")
position=-1
flag=False
n=len(s)
while True:
    position=s.find(substr,position+1,n)
    if position == -1:
        break
    print("Found at position", position)
    flag = True
if flag == False:
    print("Not Found")
```

```python
s=input("Enter the string")
words=s.split()
#print(words)
new_string=[]
i=len(words)-1
while i>=0:
    #print(i)
    new_string.append(words[i])
    i=i-1
output=' '.join(new_string)
print(output)
```

```python
s=input("Enter the string")
words=s.split()
#print(words)
new_string=[]
i=0
while i<len(words):
 #   print(i)
  #  print(words[i])
    new_string.append(words[i][::-1])
   # print(new_string)
    i=i+1
output=' '.join(new_string)
print(output)
```

```python
s=input("Enter the string")
s1=s2=output=''
for x in s:
    if x.isalpha():
        s1=s1+x
    else:
        s2=s2+x
for x in sorted(s1):
    output=output+x
for x in sorted(s2):
    output=output+x
print(output)
```
