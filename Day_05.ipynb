```python
s = input("Enter the string")
s2=[]
s3=""
for i in range(len(s)-1,-1,-1):
    s2.append(s[i])
    s3="".join(s2)
if s == s3:
    print("Given String is a palindrom",s)
else:
    print("Given String is not a Palindrom",s)
    
```
